### MSL data processing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants as c
from scipy import integrate, signal, interpolate
import requests
import holoviews as hv
hv.extension('bokeh')
import hvplot.pandas
```

```python
def smooth(data, win=11): 
    smoothed_data = signal.savgol_filter(data, win, 3)
    return smoothed_data

def integrate_mc(mc, identifier):
    '''Integrate the voltage induced in the Mirnov coils to obtain the magnetic field [mT]'''
    const = 1/(3.8e-03)
    mc -= mc.loc[:0.9e-3].mean()  #remove offset
    mc = pd.Series(integrate.cumtrapz(mc, x=mc.index, initial=0) * const,
                    index=mc.index*1000, name= 'B'+identifier[1:])    #integration
    return mc*1e3 

def integrate_sc(mc, identifier):
    '''Integrate the voltage induced in the Saddle coil to obtain the vertical component of the magnetic field [mT]'''
    const = 1/(0.83)
    mc -= mc.loc[:0.9e-3].mean()  #remove offset
    mc = pd.Series(integrate.cumtrapz(mc, x=mc.index, initial=0) * const,
                    index=mc.index*1000, name= 'B'+identifier[1:])    #integration
    return mc*1e3 

def get_closest_value(data, value, arg = False):
    if arg:
        return np.abs(data - value).argmin()
    else:
        return data[np.abs(data - value).argmin()]
    
```

### Load data from other diagnostics

```python
def load_data(shot_no):
    mirnov_url   = lambda identifier: f'http://golem.fjfi.cvut.cz/shots/{shot_no}/Diagnostics/LimiterMirnovCoils/{identifier}.csv'
    # stab_url     = lambda position: f'http://golem.fjfi.cvut.cz/shots/{shot_no}/Infrastructure/PositionStabilization/U^{position}_fg.csv' #!!! zrejme prohozene kanaly pro U <-> I!!!
    stab_url     = lambda position: f'http://golem.fjfi.cvut.cz/shots/{shot_no}/Infrastructure/PositionStabilization/U^{position}_currclamp.csv' 

    # Standard diagnostics
    BD = ['U_loop', 'Bt', 'Ich']
    df_BD = pd.concat([pd.read_csv(f'http://golem.fjfi.cvut.cz/shots/{shot_no}/Diagnostics/BasicDiagnostics/Results/{sig}.csv', 
                                names = ['Time', sig], index_col=0) for sig in BD], axis = 'columns')
    
    # Mirnov coils
    df_mirnov = pd.concat([integrate_mc(pd.read_csv(mirnov_url(sig), names = ['Time', sig], index_col = 'Time').squeeze(), sig) 
                           for sig in ['U_mc1','U_mc5','U_mc9','U_mc13']], axis = 'columns')
    
    # Currents in the stabilization coils
    cI = 1/0.05
    df_stab = pd.concat([pd.read_csv(stab_url(pos), names = [pos])*cI for pos in ['radial','vertical']], axis = 'columns')
    # get oscilloscope's time
    dt = float(requests.get(f'http://golem.fjfi.cvut.cz/shotdir/{shot_no}/Devices/Oscilloscopes/RigolMSO5204-a/ScopeSetup/XINC').text)*1e6
    df_stab.index = pd.Series(np.linspace(0, dt, df_stab.shape[0])).rename('Time')
    # smooth signals
    df_stab['radial'] = smooth(df_stab['radial'])
    df_stab['vertical'] = smooth(df_stab['vertical'])
    
    return df_BD, df_mirnov, df_stab
```

```python
def load_data_saddle(shot_no):
    saddle_url = f'http://golem.fjfi.cvut.cz/shots/0/Diagnostics/FastSpectrometry/U_HeI.csv'
    
    # Saddle coil
    df_saddle = integrate_mc(pd.read_csv(saddle_url, names = ['Time', 'B_z'], index_col = 'Time').squeeze(), 'B_z') 
    
    return df_saddle
```

### Load data measured by MSL probe

```python
def load_MSL_data(shot_no, pf, trigger_pos):
    data = pd.read_csv(f'{pf}{shot_no}.txt', skiprows = 22, sep = ';', 
                       names = ['Time [ms]', 'B_x', 'B_y', 'B_z'], 
                       decimal = ',', usecols = [0, 2, 3, 4], index_col = 0)
    data = data.replace('\t', ' ', regex=True)
    # remove offset + convert to float + remove data before trigger
    data = data.astype(np.float64)
    for cl in data.columns:
        data[cl] -= data[cl].loc[:0.5].mean()
        data[cl]  = smooth(data[cl],21)
    data.index -= trigger_pos
    return data
```

```python
pf_MSL = f'../../experiments/{date}/MSL_data/'
trigger_pos = 2.56
df_MSL = load_MSL_data(shot_no, pf_MSL, trigger_pos)
```

### Plot signals

```python
def plot_overview(df_mc, df_stab, df_BD, df_MSL, shot_no, save = False):
    fig, axs = plt.subplots(3,3, figsize = (14,12), constrained_layout=True) 
    
    # Set title
    url_comment = f'http://golem.fjfi.cvut.cz/shots/{shot_no}/Production/Parameters/comment'
    comment = requests.get(url_comment).text
    fig.suptitle(f'#{shot_no}, {comment}', fontsize = 16)
    
    # MSL data
    df_MSL['B_x'].plot(ax = axs[0,0], label = '$B_x^{MSL}$', ylabel = 'B [mT]')
    df_MSL['B_y'].plot(ax = axs[0,1], label = '$B_y^{MSL}$', ylabel = 'B [mT]')
    df_MSL['B_z'].plot(ax = axs[0,2], label = '$B_z^{MSL}$', ylabel = 'B [mT]')
    
    # Currents in stabilization coils and chamber
    df_stab['vertical'].plot(ax = axs[2,0], label = '$I_{vertstab}$', ylabel = 'I [A]')
    df_stab['radial'].plot(ax = axs[2,1], label = '$I_{horstab}$', ylabel = 'I [A]')
    (df_BD['Ich']*1e3).plot(ax = axs[2,2], label = '$I_{ch}$', ylabel = 'I [A]')
    
    # MC coils
    (df_mc[['B_mc1', 'B_mc9']].abs()).plot(ax = axs[1,0], ylabel = 'B [mT]')
    (df_mc[['B_mc5', 'B_mc13']].abs()).plot(ax = axs[1,2], ylabel = 'B [mT]')
    
    # Saturation of the iron core
    phi =  pd.Series(integrate.cumtrapz(df_BD['U_loop'], x = df_BD['U_loop'].index*1e-3, initial=0), index=df_BD['U_loop'].index, name = 'phi')
    sat = phi[phi>=0.12]
    phi.plot(ax = axs[1,1], ylabel = '$\Phi$ [Wb]', label = '$\Phi(t) = \int U_{loop} dt$')
    if sat.size>1:
        axs[1,0].axvline(sat.index[0], color = 'tab:red')
        axs[1,1].axvline(sat.index[0], color = 'tab:red')
        axs[1,2].axvline(sat.index[0], color = 'tab:red')
        axs[2,2].axvline(sat.index[0], color = 'tab:red')
        
        axs[1,1].axhline(sat.iloc[0], color = 'tab:red')
        
        axs[1,1].annotate(f't = {round(sat.index[0], 2)} ms', xy=(sat.index[0], sat.iloc[0]), xytext=(sat.index[0]+0.8, sat.iloc[0]-0.01), arrowprops=dict(arrowstyle="->"))
    
    for ax in axs.flat:
        ax.legend()
        ax.set(xlabel = 'Time [ms]', xlim = (0,30))
    
    if save:
        plt.savefig(f'figs/{shot_no}.png')
    return
```

```python
plot_overview(df_mc, df_stab, df_BD, df_MSL, shot_no)
```
